```python
#original_data[(original_data['toeflScore']>120)&(original_data['toeflScore']<=300)]
#新老gre评分标准，老GRE800分为满分，对应新GRE340,老GRE每下降10分，新GRE下降一分
def GREvscore_change(x):
    try:
        if x<=170:
            return(x)
        else:
            if x >=760:
                return(170)
            elif x in [750,740]:
                return(169)
            elif x in [730,720]:
                return(168)
            else:
                return(168-(720-x)/10)
    except:
        return(x)

    
original_data['greV'] = original_data['greV'].apply(lambda x:GREvscore_change(x))
original_data['greQ'] = original_data['greQ'].apply(lambda x:GREvscore_change(x))
```
